```
countries_df = []
years = [1990.0, 1995.0, 2000.0, 2005.0, 2010.0, 2015.0, 2017.0]

for year in years:
  df_temp = countries_migration.loc[countries_migration["Unnamed: 0"] == year]
  df_temp = df_temp.sort_values("Unnamed: 2")
  df_temp = df_temp[countries_names]
  countries_df.append(df_temp)
```
